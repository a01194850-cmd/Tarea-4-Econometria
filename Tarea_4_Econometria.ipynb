```python
# ============================================
# ANALISIS ECONOMÉTRICO INTRADÍA (TSLA, NFLX, ROKU, KO, WMT, PEP)
# Colab-ready: estacionariedad, cointegración y diagnóstico de residuos
# Librerías: pandas, numpy, matplotlib, seaborn, statsmodels, scipy
# NOTA: Opción 2 → Phillips–Perron no disponible: se usa sustituto ADF interno.
# ============================================

# ===== 0. CONFIGURACIÓN GLOBAL =====
import warnings
warnings.filterwarnings("ignore")

import itertools
import io
from typing import Dict, List, Tuple

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from IPython.display import Markdown, display

# Importar pruebas disponibles en esta versión
from statsmodels.tsa.stattools import adfuller, kpss, coint, zivot_andrews
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.vector_ar.vecm import coint_johansen
from statsmodels.graphics.gofplots import qqplot
from scipy.stats import shapiro, jarque_bera, anderson, skew, kurtosis

# === SUSTITUTO Phillips–Perron (PP) ===
# En esta Opción 2, no actualizamos statsmodels. Implementamos un wrapper
# que usa ADF como aproximación para que el pipeline no falle.
def phillips_perron(series):
    """
    Sustituto simplificado de la prueba Phillips–Perron cuando no está disponible
    en la versión instalada de statsmodels.
    Implementación: usa ADF (autolag='AIC') como aproximación.
    Devuelve (stat, pvalue, None, None) para compatibilidad con el resto del código.
    """
    s = pd.Series(series).dropna()
    stat, pvalue, _, _, _, _ = adfuller(s, autolag="AIC")
    return stat, pvalue, None, None

# Parámetros de estilo y significancia
ALPHA = 0.05
sns.set(style="whitegrid", context="notebook")
plt.rcParams["figure.figsize"] = (10, 5)
plt.rcParams["axes.titlesize"] = 13
plt.rcParams["axes.labelsize"] = 11

def md(txt: str):
    display(Markdown(txt))

md("## Análisis econométrico intradía — TSLA, NFLX, ROKU, KO, WMT, PEP\n"
   f"- Nivel de significancia: **α = {ALPHA}**\n"
   "- Evaluación de raíces unitarias (ADF, KPSS, PP≈ADF, Zivot–Andrews), cointegración (Engle–Granger y Johansen) y diagnóstico de residuos en 1min, 3min y 5min.\n"
   "- Todo se muestra en pantalla: sin exportar archivos.\n"
   "- En Colab, sube el **.xlsx** (hojas `1min`, `3min`, `5min`).\n"
   "> **Nota PP:** Phillips–Perron no está disponible en esta versión; se usa ADF como sustituto para no interrumpir el flujo.")

# ===== 1. CARGA DE DATOS =====
def load_excel_to_dict() -> Dict[str, pd.DataFrame]:
    """
    Carga un archivo Excel con hojas 1min, 3min, 5min.
    - Detecta la columna de fecha (nombre 'date' o la primera columna).
    - Convierte la fecha a datetime (maneja zona horaria si viene incluida).
    - La establece como índice y asegura orden temporal.
    """
    filename = None
    try:
        # Intento Colab
        from google.colab import files  # type: ignore
        md("**Sube el archivo .xlsx con las tres hojas (1min, 3min, 5min).**")
        uploaded = files.upload()
        filename = list(uploaded.keys())[0]
    except Exception:
        md("> *Nota:* No se detectó entorno Colab para `files.upload()`. "
           "Asegúrate de que el archivo .xlsx esté disponible en el entorno y fija el `filename` manualmente.")

    if filename is None:
        # Fallback: intenta un nombre por defecto (puedes editarlo)
        filename = "data_intradia.xlsx"

    sheet_names = ["1min", "3min", "5min"]
    data = pd.read_excel(filename, sheet_name=sheet_names)

    dfs = {}
    for sh, df in data.items():
        # Detectar columna fecha
        date_col = None
        candidates = [c for c in df.columns if str(c).strip().lower() in ("date", "fecha", "time", "datetime", "timestamp")]
        if candidates:
            date_col = candidates[0]
        else:
            date_col = df.columns[0]  # asume primera columna es fecha

        # Parseo de fecha (con o sin tz)
        dt = pd.to_datetime(df[date_col], utc=True, errors="coerce")
        df = df.drop(columns=[date_col]).copy()
        df.index = dt
        df = df.sort_index()
        # Si quieres naive: df.index = df.index.tz_convert("UTC").tz_localize(None) si fuese tz-aware
        dfs[sh] = df

    return dfs

dfs = load_excel_to_dict()

# Revisión de estructura y detección dinámica de tickers
def detect_tickers(df: pd.DataFrame) -> List[str]:
    # Filtra columnas numéricas; evita hardcodeo
    numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    return numeric_cols

md("### Vista previa de las hojas")
for freq, df in dfs.items():
    md(f"**Hoja `{freq}`** — {df.shape[0]} filas × {df.shape[1]} columnas")
    display(df.head(3))

# ===== 2. GENERADOR DE PARES =====
def all_pairs(tickers: List[str]) -> List[Tuple[str, str]]:
    return list(itertools.combinations(tickers, 2))

# ===== 3. PRUEBAS DE ESTACIONARIEDAD (ADF, KPSS, PP≈ADF, Zivot–Andrews) =====
def run_unit_root_tests(series: pd.Series, alpha: float = ALPHA) -> dict:
    """
    Ejecuta ADF, KPSS, PP≈ADF (sustituto), Zivot–Andrews sobre una serie.
    Retorna estadísticas y p-valores, más interpretaciones cortas.
    """
    out = {"n": series.dropna().shape[0]}

    # ADF: H0 = raíz unitaria (no estacionaria)
    try:
        adf_stat, adf_p, _, _, crit, _ = adfuller(series.dropna(), autolag="AIC")
        out["ADF_stat"] = adf_stat
        out["ADF_p"] = adf_p
        out["ADF_interp"] = "ADF p < α → se rechaza H₀ → la serie es **estacionaria**" if adf_p < alpha else "ADF p ≥ α → **no** se rechaza H₀ → la serie es **no estacionaria**"
    except Exception as e:
        out["ADF_error"] = str(e)

    # KPSS: H0 = estacionaria
    try:
        kpss_stat, kpss_p, _, _ = kpss(series.dropna(), regression="c", nlags="auto")
        out["KPSS_stat"] = kpss_stat
        out["KPSS_p"] = kpss_p
        out["KPSS_interp"] = "KPSS p < α → se rechaza H₀ → la serie es **no estacionaria**" if kpss_p < alpha else "KPSS p ≥ α → **no** se rechaza H₀ → la serie es **estacionaria**"
    except Exception as e:
        out["KPSS_error"] = str(e)

    # PP≈ADF (sustituto): H0 = raíz unitaria
    try:
        pp_stat, pp_p, _, _ = phillips_perron(series.dropna())
        out["PP_stat"] = pp_stat
        out["PP_p"] = pp_p
        out["PP_interp"] = "PP p < α → se rechaza H₀ → la serie es **estacionaria** (PP≈ADF)" if pp_p < alpha else "PP p ≥ α → **no** se rechaza H₀ → la serie es **no estacionaria** (PP≈ADF)"
    except Exception as e:
        out["PP_error"] = str(e)

    # Zivot–Andrews: H0 = raíz unitaria con ruptura
    try:
        za_stat, za_p, za_cp = zivot_andrews(series.dropna())
        out["ZA_stat"] = za_stat
        out["ZA_p"] = za_p
        out["ZA_break_index"] = int(za_cp)
        out["ZA_interp"] = "ZA p < α → se rechaza H₀ → **estacionaria con ruptura**" if za_p < alpha else "ZA p ≥ α → **no** se rechaza H₀ → **no estacionaria** (con posible ruptura)."
    except Exception as e:
        out["ZA_error"] = str(e)

    return out

def unit_root_panel(df: pd.DataFrame, alpha: float = ALPHA) -> pd.DataFrame:
    rows = []
    for col in detect_tickers(df):
        res = run_unit_root_tests(df[col], alpha)
        res["Ticker"] = col
        rows.append(res)
    panel = pd.DataFrame(rows).set_index("Ticker")
    return panel

md("## 1) Pruebas de estacionariedad por frecuencia")
unit_results = {}
for freq, df in dfs.items():
    md(f"### Estacionariedad — `{freq}`")
    panel = unit_root_panel(df)
    unit_results[freq] = panel
    display(panel[[c for c in panel.columns if "p" in c.lower() or "interp" in c.lower()]])

# ===== 4. ACF / PACF E INTERPRETACIÓN HEURÍSTICA =====
def acf_pacf_plots(series: pd.Series, title: str, lags: int = 40):
    fig, ax = plt.subplots(1, 1)
    series.plot(ax=ax)
    ax.set_title(f"Serie: {title}")
    ax.set_xlabel("Tiempo")
    ax.set_ylabel("Nivel")
    plt.show()

    fig = plot_acf(series.dropna(), lags=lags)
    plt.title(f"ACF — {title}")
    plt.show()

    fig = plot_pacf(series.dropna(), lags=lags, method="ywm")
    plt.title(f"PACF — {title}")
    plt.show()

def acf_pacf_interpretation(series: pd.Series) -> str:
    """
    Heurística rápida para interpretar forma de ACF/PACF:
    - Caída lenta en ACF sugiere no estacionariedad o componente AR alto.
    - Corte en PACF con 1-2 rezagos sugiere AR(p) bajo.
    - Corte en ACF y caída lenta en PACF sugiere MA(q).
    """
    msg = []
    s = series.dropna()
    if s.shape[0] < 60:
        return "Serie corta: interpretar con cautela."
    acf_vals = pd.Series([s.autocorr(lag) for lag in range(1, 6)])
    if (acf_vals.abs() > 0.5).any():
        msg.append("ACF alta en rezagos iniciales → posible **no estacionariedad** o fuerte persistencia.")
    elif (acf_vals.abs() > 0.2).any():
        msg.append("ACF moderada → posible estructura ARMA de bajo orden.")
    else:
        msg.append("ACF baja → señales de **estacionariedad** (al menos a corto plazo).")
    return " ".join(msg)

md("## 2) ACF / PACF y patrones")
for freq, df in dfs.items():
    md(f"### `{freq}` — patrones ACF/PACF")
    for col in detect_tickers(df):
        md(f"**{col}** — {acf_pacf_interpretation(df[col])}")
        acf_pacf_plots(df[col], title=f"{col} ({freq})", lags=40)

# ===== 5. COINTEGRACIÓN DE ENGLE–GRANGER (TODOS LOS PARES) =====
def engle_granger_matrix(df: pd.DataFrame, alpha: float = ALPHA) -> pd.DataFrame:
    tickers = detect_tickers(df)
    pairs = all_pairs(tickers)
    rows = []
    for a, b in pairs:
        try:
            stat, p, _ = coint(df[a].dropna(), df[b].dropna())
            rows.append({"A": a, "B": b, "EG_stat": stat, "EG_p": p, "Cointegrado": p < alpha})
        except Exception:
            rows.append({"A": a, "B": b, "EG_stat": np.nan, "EG_p": np.nan, "Cointegrado": False})
    out = pd.DataFrame(rows).sort_values("EG_p")
    return out

md("## 3) Cointegración (Engle–Granger) — tabla por frecuencia")
eg_results = {}
for freq, df in dfs.items():
    md(f"### `{freq}`")
    eg_table = engle_granger_matrix(df)
    eg_results[freq] = eg_table
    display(
        eg_table.style
        .format({"EG_p": "{:.4f}", "EG_stat": "{:.3f}"})
        .apply(lambda s: ["background-color: #c7f7c1" if (s.name and eg_table.loc[s.name, "Cointegrado"]) else "" for _ in s], axis=1)
    )
    n_c = eg_table["Cointegrado"].sum()
    md(f"- **Pares cointegrados (p < {ALPHA}):** {int(n_c)} de {eg_table.shape[0]} pares.")

# ===== 6. COINTEGRACIÓN MULTIVARIANTE (JOHANSEN) =====
def johansen_summary(df: pd.DataFrame, det_order: int = 0, k_ar_diff: int = 1, alpha: float = ALPHA) -> pd.DataFrame:
    """
    Ejecuta prueba de Johansen para todas las columnas numéricas.
    det_order: 0 = sin determinísticos, 1 = const, etc.
    k_ar_diff: número de rezagos en diferencias (no confundir con VAR lags).
    Retorna tabla con estadísticas trace y maxeig vs críticos, y rango estimado.
    """
    X = df[detect_tickers(df)].dropna().values
    res = coint_johansen(X, det_order, k_ar_diff)

    ranks = list(range(X.shape[1]))  # 0..k-1
    out_rows = []
    for r in ranks:
        out_rows.append({
            "rank_r": r,
            "trace_stat": res.lr1[r],
            "trace_crit_95": res.cvt[r, 1],
            "trace_H0_reject@5%": res.lr1[r] > res.cvt[r, 1],
            "maxeig_stat": res.lr2[r],
            "maxeig_crit_95": res.cvm[r, 1],
            "maxeig_H0_reject@5%": res.lr2[r] > res.cvm[r, 1],
        })
    out = pd.DataFrame(out_rows)

    # Rango estimado: mayor r tal que se rechaza H0 por trace a 5%
    rank_est = out.loc[out["trace_H0_reject@5%"]].rank_r.max() if out["trace_H0_reject@5%"].any() else 0
    md(f"- **Johansen (det_order={det_order}, k_ar_diff={k_ar_diff})** → rango estimado: **{int(rank_est)}** "
       f"(número de relaciones cointegrantes).")

    return out

md("## 4) Cointegración multivariante (Johansen) — resumen por frecuencia")
johansen_results = {}
for freq, df in dfs.items():
    md(f"### `{freq}`")
    try:
        joh = johansen_summary(df, det_order=0, k_ar_diff=1)
        johansen_results[freq] = joh
        display(joh)
    except Exception as e:
        md(f"> No fue posible ejecutar Johansen en `{freq}`: {e}")

# ===== 7. VISUALIZACIÓN DE LAS RELACIONES MÁS COINTEGRADAS + RESIDUOS =====
def normalize(s: pd.Series) -> pd.Series:
    return (s - s.mean()) / s.std(ddof=0)

def plot_top_coint_pairs(df: pd.DataFrame, eg_table: pd.DataFrame, top_n: int = 3):
    table = eg_table.dropna(subset=["EG_p"]).sort_values("EG_p").head(top_n)
    for _, row in table.iterrows():
        a, b = row["A"], row["B"]
        s1, s2 = df[a].dropna(), df[b].dropna()
        common_idx = s1.index.intersection(s2.index)
        s1, s2 = s1.loc[common_idx], s2.loc[common_idx]

        md(f"#### Par {a} ~ {b} — EG p = {row['EG_p']:.4f} ({'cointegrado' if row['Cointegrado'] else 'no cointegrado'})")
        # Series normalizadas
        plt.figure()
        normalize(s1).plot(label=a)
        normalize(s2).plot(label=b)
        plt.title(f"Series normalizadas: {a} vs {b}")
        plt.legend()
        plt.xlabel("Tiempo"); plt.ylabel("Z-score")
        plt.show()

        # Regresión OLS simple para extraer residuo (Engle–Granger)
        X = np.vstack([np.ones(len(s1)), s2.values]).T
        beta = np.linalg.lstsq(X, s1.values, rcond=None)[0]
        resid = s1.values - X @ beta

        # Residuo + ADF sobre residuo
        plt.figure()
        pd.Series(resid, index=common_idx).plot()
        plt.title(f"Residuo de {a} sobre {b} (Engle–Granger)")
        plt.xlabel("Tiempo"); plt.ylabel("Residuo")
        plt.show()

        try:
            adf_stat, adf_p, *_ = adfuller(resid, autolag="AIC")
            md(f"- **ADF(residuo)** p = {adf_p:.4f} → "
               + ("se rechaza H₀ → residuo **estacionario** → cointegración confirmada."
                  if adf_p < ALPHA else
                  "no se rechaza H₀ → residuo **no** estacionario."))
        except Exception as e:
            md(f"> ADF en residuo no disponible: {e}")

md("## 5) Visualización de las relaciones más cointegradas (por frecuencia)")
for freq, df in dfs.items():
    if freq in eg_results:
        md(f"### `{freq}`")
        plot_top_coint_pairs(df, eg_results[freq], top_n=3)

# ===== 8. DIAGNÓSTICO DE RESIDUOS (Normalidad, simetría, colas) =====
def residual_diagnostics(resid: pd.Series, title: str):
    r = resid.dropna()
    md(f"**Diagnóstico de residuos — {title}**")

    # Histograma + KDE
    plt.figure()
    sns.histplot(r, kde=True, stat="density")
    plt.title(f"Histograma + KDE — {title}")
    plt.xlabel("Residuo"); plt.ylabel("Densidad")
    plt.show()

    # QQ plot
    qqplot(r, line="s")
    plt.title(f"QQ plot — {title}")
    plt.show()

    # Estadísticos y pruebas
    sk = skew(r, bias=False)
    ku = kurtosis(r, fisher=True, bias=False)

    try:
        sh_stat, sh_p = shapiro(r if len(r) <= 5000 else r.sample(5000, random_state=42))
        sh_msg = f"Shapiro–Wilk p = {sh_p:.4f} → " + ("**normalidad no rechazada**" if sh_p >= ALPHA else "**normalidad rechazada**")
    except Exception as e:
        sh_stat, sh_p, sh_msg = np.nan, np.nan, f"Shapiro no disponible: {e}"

    try:
        jb_stat, jb_p = jarque_bera(r)
        jb_msg = f"Jarque–Bera p = {jb_p:.4f} → " + ("**normalidad no rechazada**" if jb_p >= ALPHA else "**normalidad rechazada**")
    except Exception as e:
        jb_stat, jb_p, jb_msg = np.nan, np.nan, f"Jarque–Bera no disponible: {e}"

    try:
        ad = anderson(r, dist="norm")
        # anderson no da p-valor; comparamos estadístico con crítico 5%
        ad_stat = ad.statistic
        crit_levels = list(ad.significance_level)
        if 5.0 in crit_levels:
            crit_5 = ad.critical_values[crit_levels.index(5.0)]
            ad_msg = f"Anderson–Darling stat = {ad_stat:.3f} vs crit(5%) = {crit_5:.3f} → " + ("**normalidad rechazada**" if ad_stat > crit_5 else "**normalidad no rechazada**")
        else:
            ad_msg = f"Anderson–Darling stat = {ad_stat:.3f} (sin crítico 5% disponible)"
    except Exception as e:
        ad_msg = f"Anderson–Darling no disponible: {e}"

    md(f"- Asimetría (skew): **{sk:.3f}**, Curtosis (exceso): **{ku:.3f}**")
    md(f"- {sh_msg}")
    md(f"- {jb_msg}")
    md(f"- {ad_msg}")

# Diagnóstico en el mejor par cointegrado por frecuencia
md("## 6) Diagnóstico de residuos en el mejor par cointegrado (por frecuencia)")
for freq, df in dfs.items():
    eg = eg_results.get(freq)
    if eg is None or eg.dropna(subset=["EG_p"]).empty:
        md(f"### `{freq}` — Sin resultados EG válidos.")
        continue
    best = eg.dropna(subset=["EG_p"]).sort_values("EG_p").iloc[0]
    a, b = best["A"], best["B"]
    s1, s2 = df[a].dropna(), df[b].dropna()
    idx = s1.index.intersection(s2.index)
    s1, s2 = s1.loc[idx], s2.loc[idx]
    X = np.vstack([np.ones(len(s1)), s2.values]).T
    beta = np.linalg.lstsq(X, s1.values, rcond=None)[0]
    resid = pd.Series(s1.values - X @ beta, index=idx)
    md(f"### `{freq}` — Residuo de **{a}** sobre **{b}** (par con menor p)")
    residual_diagnostics(resid, title=f"{a}~{b} ({freq})")

# ===== 9. PIPELINE OPCIONAL (Ejecución total) =====
def full_pipeline(dfs: Dict[str, pd.DataFrame], alpha: float = ALPHA):
    """
    Ejecuta: pruebas de raíz unitaria, EG en pares, Johansen, visualizaciones top pares y diagnóstico de residuos.
    Útil si quieres re-ejecutar todo con un click.
    """
    md("# ▶️ Pipeline general")
    for freq, df in dfs.items():
        md(f"## Frecuencia `{freq}`")

        # 1) Estacionariedad
        md("### Estacionariedad (ADF, KPSS, PP≈ADF, ZA)")
        panel = unit_root_panel(df, alpha)
        display(panel[[c for c in panel.columns if "p" in c.lower() or "interp" in c.lower()]])

        # 2) ACF/PACF resumen
        md("### ACF/PACF — interpretación heurística")
        for col in detect_tickers(df):
            md(f"- **{col}**: {acf_pacf_interpretation(df[col])}")

        # 3) Engle–Granger
        md("### Engle–Granger (todos los pares)")
        eg_table = engle_granger_matrix(df, alpha)
        display(eg_table.style.format({"EG_p": "{:.4f}", "EG_stat": "{:.3f}"})
                .apply(lambda s: ["background-color: #c7f7c1" if (s.name and eg_table.loc[s.name, "Cointegrado"]) else "" for _ in s], axis=1))

        # 4) Johansen
        md("### Johansen (multivariado)")
        try:
            joh = johansen_summary(df, det_order=0, k_ar_diff=1, alpha=alpha)
            display(joh)
        except Exception as e:
            md(f"> Johansen no disponible: {e}")

        # 5) Visualización top3 pares + ADF del residuo
        md("### Visualización — Top 3 pares más cointegrados")
        plot_top_coint_pairs(df, eg_table, top_n=3)

        # 6) Diagnóstico de residuos en el top1
        if not eg_table.dropna(subset=["EG_p"]).empty:
            best = eg_table.dropna(subset=["EG_p"]).sort_values("EG_p").iloc[0]
            a, b = best["A"], best["B"]
            s1, s2 = df[a].dropna(), df[b].dropna()
            idx = s1.index.intersection(s2.index)
            s1, s2 = s1.loc[idx], s2.loc[idx]
            X = np.vstack([np.ones(len(s1)), s2.values]).T
            beta = np.linalg.lstsq(X, s1.values, rcond=None)[0]
            resid = pd.Series(s1.values - X @ beta, index=idx)
            md(f"### Diagnóstico de residuos — mejor par: **{a}~{b}**")
            residual_diagnostics(resid, title=f"{a}~{b} ({freq})")

md("---\n**Sugerencia:** Si quieres ejecutar **todo** en orden con un clic, corre: `full_pipeline(dfs)` en una celda aparte.\n"
   "Puedes ajustar `det_order` y `k_ar_diff` en Johansen según tu juicio económico y la frecuencia intradía.")

# FIN DEL CUADERNO
```
